```python
while True:
    cv2.imshow('result',img)
#     27 is ASCII value of esc key(i.e window will be terminated only when we press esc key)
    if cv2.waitKey(2) == 27:
        break
cv2.destroyAllWindows()
```

```python
haar_data = cv2.CascadeClassifier('haarCascadeFrontalFace.xml')
```

```python
haar_data.detectMultiScale(img)
```

```python
while True:
    faces = haar_data.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),4)
    cv2.imshow('result',img)
#     27 is ASCII value of esc key(i.e window will be terminated only when we press esc key)
    if cv2.waitKey(2) == 27:
        break
cv2.destroyAllWindows()
```

```python
capture = cv2.VideoCapture(0)
data = []
while True:
    flag, img = capture.read()
    if(flag):
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),4)
            face = img[y:y+h,x:x+w, :]
            face = cv2.resize(face,(50,50))
            data.append(face)
            print(len(data))
            if(len(data)<2000):
                data.append(face)
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27 or len(data)>=1000:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
```

```python
np.save('without_mask.npy',data)
```

```python
np.save('with_mask.npy',data)
```
