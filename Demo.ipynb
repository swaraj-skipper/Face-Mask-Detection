```python
import numpy as np 
```

```python
import cv2
```

```python
from sklearn.decomposition import PCA
```

```python
with_mask = np.load('with_mask.npy')
```

```python
without_mask = np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
with_mask = with_mask.reshape(1000,50*50*3)
```

```python
without_mask = without_mask.reshape(1000,50*50*3)
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
x = np.r_[with_mask,without_mask]
```

```python
x.shape
```

```python
labels = np.zeros(x.shape[0])
```

```python
labels[1000:] = 1.0
names = {0:'MASK', 1:'NO MASK'}
```

```python
#1 -- without mask, 0 -- with mask
```

```python
#svm - support vector machine
#SVC - support vector classification
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,labels,test_size = 0.20)
```

```python
x_train.shape
```

```python
pca = PCA(n_components = 3)
x_train = pca.fit_transform(x_train)
```

```python
x_train[0]
```

```python
x_train.shape
```

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
x_test = pca.transform(x_test)
y_pred = svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
haar_data = cv2.CascadeClassifier('haarCascadeFrontalFace.xml')
capture = cv2.VideoCapture(0)
data = []
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    flag, img = capture.read()
    
    if(flag):
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),4)
            face = img[y:y+h,x:x+w, :]
            face = cv2.resize(face,(50,50))
            face = face.reshape(1,-1)
            face = pca.transform(face)
            pred = svm.predict(face)
            n = names[int(pred)]
            cv2.putText(img, n,(x,y),font,1,(255,0,0),2)
            print(n)
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27:
            break
capture.release()
cv2.destroyAllWindows()
```
